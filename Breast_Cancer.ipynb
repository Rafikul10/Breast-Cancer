# Train Test dataset Ssplit

```python
#from sklearn.model_selection import train_test_split
x=data.iloc[:,:]
y=data.iloc[:,-1].copy()
```

```python
x.drop(columns={"Status","Grade"},axis=1,inplace=True)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.shape
```

```python
y_test.head()
```

# Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train.shape
```

```python
x_test
```


# Model Training

```python
#from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
#from sklearn.naive_bayes import GaussianNB
#model = GaussianNB()
#from sklearn.svm import SVC
#from sklearn.neighbors import KNeighborsClassifier
#from sklearn.tree import DecisionTreeClassifier
#model=LogisticRegression()
#model=KNeighborsClassifier()
model=RandomForestClassifier()
#model=SVC()
#model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

# Coefficient and Intercept

```python
#print(model.coef_)
#print(model.intercept_)
```

```python
#model.get_params()
```

# Prediction With Test Data

```python
reg_predict=model.predict(x_test)
```

```python
reg_predict
```

```python
np.array(y_test)
```

```python
plt.scatter(y_test,reg_predict)
```

```python
residuals=y_test-reg_predict
```

```python
sns.displot(residuals,kind="kde")
```

# Error Calculations

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_absolute_error(y_test,reg_predict))
print(mean_squared_error(y_test,reg_predict))
print(np.sqrt(mean_squared_error(y_test,reg_predict)))
```
